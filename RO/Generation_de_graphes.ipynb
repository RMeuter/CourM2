```python
#creation d'un graphe aleatoire (pas forcement connecte)

from sage.graphs.graph_generators_pyx import RandomGNP
set_random_seed(0)

# on cree un graphe de 50 sommets.
# la probabilité que deux sommets quelconques soient connectés est de 20%
G=RandomGNP(50, .2, directed=False)
G.show()

# on cree un second graphe de 100 sommets.
# la probabilité que deux sommets quelconques soient connectés est de 10%
G=RandomGNP(100, .1, directed=False)
G.show()

```

```python
# les graphes générés n'ont pas de capacités sur les aretes
#ajout d'une étiquette aléatoire (capacites) sur chaque arete

#génération d'un graphe aléatoire de 20 sommets, avec degré 3
G = graphs.RandomRegular(3, 20)
    
for e in G.edges() :
    x=e[0]; # x est le premiere extremite
    y=e[1]; # y est la seconde extremite
    w = randint(1, 20); #on genere un nombre aleatoire 
    G.delete_edge(x,y) #on supprime l'arete
    G.add_edge(x,y,w); # on la recrée avec une etiquette
    
G.show(edge_labels=True)

 

```
